_To do this, I will be comparing the reviews of games from GameSpot to get a general consensus on how people feel about old versus new video games._

I will start off by importing the tools that I need to complete this analysis.

```python
import pandas as pd
import matplotlib as plt
import json
import urllib
import requests
```

I need to retrieve the API keys from GameSpot.

```python
api_key = pd.read_csv(r"C:\Users\Alex Buckler\data_SKOOBY\GamesKey.txt")
key_s = api_key['GamesKey'].iloc[0]
```

```python
header = {'User-agent': 'Mozilla/5.0'}
reviews_e = 'http://www.gamespot.com/api/reviews/'
url_key = '?api_key=' + key_s
url_format = '&format=json'
url_sort = '&sort=score:desc'
url_filter ='&filter=update_date:2000-01-01|2024-11-26'
```

```python
response1 = requests.get(reviews_e + url_key + url_format + url_sort + url_filter, headers = header)
response1.status_code
```

```python
response_dict = json.loads(response1.text)
response_dict.keys()
```

```python
response_df = pd.DataFrame(response_dict['results'])
response_df.head()
```

```python
tidy_response = response_df.drop(columns = ['review_type', 'deck', 'body', 'lede', 'movie', 'image', 'id', 'videos_api_url', 'site_detail_url', 'television'])
tidy_response
```

```python
tidy_response.groupby('title').count().sort_values('score', ascending = False)
```
